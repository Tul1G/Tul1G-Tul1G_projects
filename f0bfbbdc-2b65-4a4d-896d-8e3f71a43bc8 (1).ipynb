### 2.3. Наличие пропусков в данных

- Посчитайте количество пропусков в каждом столбце в абсолютных и относительных значениях.


```python
#всего пропусков
df.isna().sum()
```

- Пропуски

В наборе данных присутствуют пропуски:

Name: 2 пропуска (0.012% от общего объема).

Year of Release: 275 пропусков (1.62%).

Genre: 2 пропуска (0.012%).

Critic Score: 8714 пропусков (51.39%).

User Score: 6804 пропуска (40.13%).

Rating: 6871 пропусков (40.52%).

```python
sales_columns = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']

# Заполнение заглушками
df['critic_score'].fillna(-1, inplace=True)
df['user_score'].fillna(-1, inplace=True)
df['rating'].fillna('non_rating', inplace=True)

# Удаление строк с пропусками в year_of_release и genre
df.dropna(subset=['year_of_release'], inplace=True)
df.dropna(subset=['genre'], inplace=True)

# Преобразование типов данных для столбцов с продажами
for column in sales_columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Заполнение пропусков в столбцах с продажами на основе среднего значения по платформе и году выхода
for column in sales_columns:
    df[column].fillna(
        df.groupby(['platform', 'year_of_release'])[column].transform('mean'),
        inplace=True
    )

# Проверка на наличие NaN значений после заполнения
print(df[['critic_score', 'user_score', 'rating', 'year_of_release', 'genre']].isnull().sum())
print(df[sales_columns].isnull().sum()) 
```

- Для столбцов critic_score , и user_score  заменили пропуски заглушкой(-1) чтобы явно обозначить отсутствие данных и избежать неверных выводов.
- Для столбца rating заменили на заглушку (например, non_rating) более точно отражает реальное отсутствие данных.
- Для стобца year_of_release удалили пропуски т.к их было мало.  
- обработали пропуски в столбцах с продажами, заменили на среднее значение  в зависимости от названия платформы и года выхода игры.

## 3. Фильтрация данных

Коллеги хотят изучить историю продаж игр в начале XXI века, и их интересует период с 2000 по 2013 год включительно. Отберите данные по этому показателю. Сохраните новый срез данных в отдельном датафрейме, например `df_actual`.

```python
#Фильтрация данных за 2000-2013 годы
df_actual = df[(df['year_of_release'] >= 2000) & (df['year_of_release'] <= 2013)].copy()

print(f"Всего игр за период 2000-2013: {len(df_actual)}")
display(df_actual.head())

# Подсчет количества игр по годам
games_per_year = df_actual['year_of_release'].value_counts().sort_index()

# Вывод количества игр по годам
print("\nКоличество игр по годам:")
print(games_per_year)
```

## 4. Категоризация данных
    
Проведите категоризацию данных:
- Разделите все игры по оценкам пользователей и выделите такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).

```python
def categorize_user_score(score):
    if pd.isna(score):
        return 'Не оценено'
    if score >= 8:
        return 'Высокая оценка (8-10)'
    elif score >= 3:
        return 'Средняя оценка (3-8)'
    else:
        return 'Низкая оценка (0-3)'

df_actual['user_score_category'] = df_actual['user_score'].apply(categorize_user_score)
```

- Разделите все игры по оценкам критиков и выделите такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).

```python
def categorize_critic_score(score):
    if pd.isna(score):
        return 'Не оценено'
    if score >= 80:
        return 'Высокая оценка (80-100)'
    elif score >= 30:
        return 'Средняя оценка (30-80)'
    else:
        return 'Низкая оценка (0-30)'

df_actual['critic_score_category'] = df_actual['critic_score'].apply(categorize_critic_score)
```

<b>Комментарий ревьюера:</b> 
    
Ты верно задаешь условия для категорий оценок. Хорошо, что ты используешь .apply(). Это стандартный подход, когда нужно создать новый столбец на основе условий.
</div>

- Выделите топ-7 платформ по количеству игр, выпущенных за весь актуальный период.

```python
#Топ-7 платформ по количеству игр
top_platforms = df_actual['platform'].value_counts().head(7)
print("\nТоп-7 платформ по количеству игр (2000-2013):")
display(top_platforms)

#Анализ распределения оценок
print("\nРаспределение оценок пользователей:")
display(df_actual['user_score_category'].value_counts())

print("\nРаспределение оценок критиков:")
display(df_actual['critic_score_category'].value_counts())

#Сохранение обработанных данных
df_actual.to_csv('games_2000_2013_processed.csv', index=False)
```
